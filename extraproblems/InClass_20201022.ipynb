## Palindrome

Write a program that checks whether an integer is a palindrome.

For eample, 121 and 131 are palindromes. However, 123 is not a palindrome.

```python
def palindrome(x):

    if not isinstance(x,str):
        x = str(x)

    return (x[::-1]==x)
```

```python
palindrome(121)
```

```python
palindrome(123)
```

Note: 
- If you want to select every k-th element of a list named x, you can use  x[::k].
- If you want to select a list starting from k-th element, you can use x[k::].
- If you want to select a list ending at k-th element, you can use x[:k:].

## Does Any Two Numbers Sum to K?

Given a list of numbers, return whether any 2 numbers sums to k.

For example, given [1, 2, 3, 5] and k of 7, return True. If k is 9, return False

```python
import numpy as np
import itertools
```

```python
def fun(x,k):
    comb_array = np.array(list(itertools.combinations(x,2)))
    comb_sum = np.sum(comb_array,1)
    return (k in comb_sum)
```

```python
fun([1,2,3,4,5],4)
```

```python
fun([1,54657,34235],2)
```

## $\pi$

```python
n=10000
circle = [0]*n
x = np.random.uniform(low=0,high=1,size=n)
y = np.random.uniform(low=0,high=1,size=n)

for i in range(n):
    hypotenuse = np.sqrt(x[i]**2+y[i]**2)
    if (hypotenuse<=1):
        circle[i]=1
(np.sum(circle)/n)*4
```
