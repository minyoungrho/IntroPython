Let's work with finance data of a company IBM.

```python
import json
from urllib.request import urlopen
```

```python
IBMdaily_api = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&apikey=demo"
```

```python
with urlopen(IBMdaily_api) as response:
    source = response.read()
```

```python
data = json.loads(source)
```

```python
type(data)
```

```python
isinstance(data,dict)
```

```python
type(data['Time Series (Daily)'])
```

```python
data['Time Series (Daily)']
```

Unlike the example that we have discussed in class, type(data['Time Series (Daily)']) isn't a list, but a dictionary.

items() returns a list of dictionary's (key,value) tuple pairs. 

For example:

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = car.items()

print(x)
```

```python
data1 = data['Time Series (Daily)']
```

```python
type(data1)
```

```python
data2 = data1.items()
```

```python
data2
```

```python
type(data2)
```

## Option 1: List

```python
date_list = []
open_list = []
high_list = []
low_list = []
close_list = []
vol_list = []

for date, price in data2:
    date_list.append(date)
    open_list.append(price['1. open'])
    high_list.append(price['2. high'])
    low_list.append(price['3. low'])
    close_list.append(price['4. close'])
    vol_list.append(price['5. volume'])    
```

```python
list(map(int, vol_list))
```

## Option 2: Dataframes

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=['date','open','high','low','close','volume'])
```

```python
df.head()
```

```python
for date, price in data2:
    row = [date,price['1. open'],price['2. high'],price['3. low'],price['4. close'],price['5. volume']]
    df.loc[len(df)] = row
```
